```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("/Users/gorkaetxebarriaatutxa/Documents/PBL4/Estadistica/Datos.csv",delimiter=';')
print(data.head())
```

```python
plt.scatter(data['incomes'],data['wasted'])
plt.xlabel('incomes')
plt.ylabel('wasted')
plt.title('incomes vs wasted')
```

```python
x=data['incomes']
y=data['wasted']
n = np.size(x)
m_x = np.mean(x)
m_y = np.mean(y)
SS_xy = np.sum(y*x) - n*m_y*m_x
SS_xx = np.sum(x*x) - n*m_x*m_x
b_1 = SS_xy / SS_xx
b_0 = m_y - b_1*m_x
SS_yy= 0
for dat in data['wasted']:
    SS_yy += (dat - m_y)**2
print('n= ',n)
print('mx= ',m_x)
print('my= ',m_y)
print('Sxy= '+str(SS_xy))
print('Sxx= '+str(SS_xx))
print('Syy= '+str(SS_yy))
print('b1= '+str(b_1))
print('b0= '+str(b_0))
```

```python
#coeficiente de correlacion
coe=data['incomes'].corr(data['wasted'], method='pearson')
print('coeficiente de correlacion= '+ str(coe))
```

```python
recta_regresion_muestral_y = b_0 + b_1 * data['incomes']
plt.plot(data['incomes'], recta_regresion_muestral_y,'ro')
plt.scatter(data['incomes'], data['wasted'])
plt.xlabel('incomes')
plt.ylabel('wasted')
plt.title('incomes vs wasted')
```
